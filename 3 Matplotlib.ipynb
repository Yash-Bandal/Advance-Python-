#32) pie with labels and colrs () =>

```python
y = np.array([35, 25, 25, 15])
labels = ["Category A", "Category B", "Category C", "Category D"]
colors = ["red", "yellow", "green", "blue"]
plt.pie(y, labels=labels, colors=colors) # from labels variable we call labels function same for colors
plt.show()
```

#34)starangle () =>

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
data = sns.load_dataset('tips')

# 1. Scatter Plot
sns.scatterplot(data=data, x='total_bill', y='tip')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()

# 2. Line Plot
sns.lineplot(data=data, x='size', y='tip')
plt.title('Line Plot of Size vs Tip')
plt.show()

# 3. Bar Plot
sns.barplot(data=data, x='day', y='total_bill')
plt.title('Bar Plot of Total Bill by Day')
plt.show()

# 4. Histogram
sns.histplot(data=data, x='total_bill', bins=20)
plt.title('Histogram of Total Bill')
plt.show()

# 5. Box Plot
sns.boxplot(data=data, x='day', y='total_bill')
plt.title('Box Plot of Total Bill by Day')
plt.show()

# 6. Violin Plot
sns.violinplot(data=data, x='day', y='total_bill')
plt.title('Violin Plot of Total Bill by Day')
plt.show()

# 7. Strip Plot
sns.stripplot(data=data, x='day', y='total_bill', jitter=True)
plt.title('Strip Plot of Total Bill by Day')
plt.show()

# 8. Swarm Plot
sns.swarmplot(data=data, x='day', y='total_bill')
plt.title('Swarm Plot of Total Bill by Day')
plt.show()

# 9. KDE Plot
sns.kdeplot(data=data, x='total_bill')
plt.title('KDE Plot of Total Bill')
plt.show()

# 10. ECDF Plot
sns.ecdfplot(data=data, x='total_bill')
plt.title('ECDF Plot of Total Bill')
plt.show()
```
