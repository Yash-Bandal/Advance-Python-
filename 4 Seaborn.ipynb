#9) Load the required Dataset:=
#tips
Loads the tips dataset, which contains information about tips received by waitstaff in a restaurant, including the total bill, tip amount, sex of the bill payer, whether they are a smoker, day of the week, time of day, and the size of the party.
#iris

#44)set () =>
-This function sets the aesthetic parameters for Seaborn plots.

```python
#style="darkgrid": Applies the "darkgrid" style, which adds a dark grid background to the plots.
#palette="muted": Sets the color palette to "muted," which provides a set of subdued colors. This palette is useful for a more understated color scheme in plots

sns.set(style="darkgrid", palette="muted")

# Plot with custom settings
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Scatterplot with Custom Settings")
plt.show()

# Reset to default Seaborn settings
sns.reset_defaults()

# Plot with default settings
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Scatterplot with Default Settings")
plt.show()
```

#45)reset_orig() =>
 - Reset the Matplotlib Configuration to Original Values
- Sets the aesthetic parameters for Seaborn plots.

```python
# Apply custom settings
sns.set(style="whitegrid", palette="dark")

# Plot with custom settings
sns.boxplot(x="species", y="sepal_length", data=iris)
plt.title("Boxplot with Custom Settings")
plt.show()

# Reset Matplotlib to its original configuration
sns.reset_orig()

# Plot with original Matplotlib settings
sns.boxplot(x="species", y="sepal_length", data=iris)
plt.title("Boxplot with Original Matplotlib Settings")
plt.show()
```
