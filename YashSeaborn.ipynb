### **Seaborn - Function Reference**

Seaborn provides a rich collection of functions to create various types of plots and visualizations. The Function Reference serves as a comprehensive guide to these functions, helping you understand how to use them effectively for different types of data analysis and visualization.

#### **1. Plotting Functions**

Seaborns plotting functions can be broadly categorized based on the type of data visualization they produce. Below are some of the most commonly used functions:

- **Relational Plots**
  - `sns.relplot()`: Combines `scatterplot()` and `lineplot()` under a common interface to visualize relationships between variables.
  - `sns.scatterplot()`: Creates scatter plots showing the relationship between two variables.
  - `sns.lineplot()`: Creates line plots to show trends over time or ordered categories.

- **Categorical Plots**
  - `sns.catplot()`: Combines several categorical plots like `stripplot`, `swarmplot`, `boxplot`, `violinplot`, and `pointplot` under a single interface.
  - `sns.stripplot()`: Creates a scatter plot where one variable is categorical.
  - `sns.boxplot()`: Creates box plots to visualize the distribution of data across different categories.

- **Distribution Plots**
  - `sns.distplot()`: (Deprecated) Previously used to plot univariate distributions.
  - `sns.histplot()`: Plots histograms and kernel density estimates.
  - `sns.kdeplot()`: Plots kernel density estimates to visualize the distribution of data.

- **Matrix Plots**
  - `sns.heatmap()`: Creates heatmaps to visualize matrix-like data, such as correlation matrices.
  - `sns.clustermap()`: Plots clustered heatmaps that group similar data points together.

- **Regression Plots**
  - `sns.lmplot()`: Visualizes linear regression models.
  - `sns.regplot()`: Fits and plots a simple linear regression model.
  - `sns.residplot()`: Plots residuals of a linear regression to check its fit.

#### **2. Styling and Aesthetics Functions**

Seaborn offers functions to customize the aesthetics of your plots, making them more informative and visually appealing.

- **Color Palettes**
  - `sns.color_palette()`: Returns a list of colors defining a color palette.
  - `sns.set_palette()`: Sets the color palette for the plots.
  - `sns.palplot()`: Visualizes the colors in a palette.

- **Figure Aesthetics**
  - `sns.set()`: Sets the aesthetic parameters in one step.
  - `sns.set_style()`: Sets the style of the plots (e.g., "white", "dark", "whitegrid", "darkgrid", "ticks").
  - `sns.set_context()`: Sets the context of the plot (e.g., "paper", "notebook", "talk", "poster").

#### **3. Utility Functions**

Seaborn also includes several utility functions that help in customizing and manipulating plots more efficiently.

- **Grids**
  - `sns.FacetGrid()`: Creates grids of subplots for visualizing complex datasets with multiple categorical variables.
  - `sns.PairGrid()`: Plots pairwise relationships between variables in a grid format.

- **Plotting Multi-Panel Data**
  - `sns.pairplot()`: Plots pairwise relationships in a dataset, similar to `PairGrid` but with a simpler interface.
  - `sns.jointplot()`: Plots the relationship between two variables along with their marginal distributions.

- **Data Preparation**
  - `sns.load_dataset()`: Loads example datasets provided by Seaborn for practice.
  - `sns.cubehelix_palette()`: Generates a color palette that is perceptually uniform.

#### **4. Advanced Functions**

For users who need more control or want to extend Seaborn's capabilities, the library offers advanced functions:

- **Custom Plots**
  - `sns.violinplot()`: Combines box plots and kernel density estimates to show distributions.
  - `sns.pointplot()`: Shows point estimates and confidence intervals using scatter plots.

- **Customizing Legends**
  - `sns.move_legend()`: Adjusts the position of legends in a plot.
  - `sns.despine()`: Removes the top and right spines from plots for a cleaner look.

- **Saving Figures**
  - `plt.savefig()`: Saves the current figure to a file, including various formats like PNG, SVG, PDF, etc.

#### **5. Examples and Documentation**
